```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_palette("gist_rainbow_r")
#sns.set_palette("Set1")
```

```python
bird= sns.load_dataset("penguins")
g = sns.FacetGrid(bird, col="island", hue="species")
g.map(plt.scatter, "flipper_length_mm", "body_mass_g", alpha=.6)
g.add_legend()
plt.show()
```

```python
sns.stripplot(x="island", y="body_mass_g", hue= "species", data=bird)
plt.show()
```

```python
sns.swarmplot(x="island", y="body_mass_g", hue="species",data=bird, dodge=True)
plt.show()
```

```python
sns.lmplot(x="body_mass_g", y="bill_length_mm", hue="sex", col="island", markers=["o", "x"], data=bird)
plt.show()
```

```python
sns.lmplot(x="body_mass_g", y="bill_length_mm", hue="sex", col="island",row="sex",order=2, markers=["o", "x"], data=bird)
plt.show()
```

```python
sns.jointplot(x="body_mass_g", y="bill_length_mm", kind="hex", data=bird )
plt.show()
```

```python
g = sns.jointplot(x="body_mass_g", y="bill_length_mm", data=bird, kind="hex", color="c")
g.plot_joint(plt.scatter, c="w", s=30, linewidth=1, marker="+")
g.set_axis_labels("Body Mass (in gram)", "Bill Length ( in mm)")
plt.show()
```

```python
sns.pairplot(bird, hue="island") 
plt.show()
```
